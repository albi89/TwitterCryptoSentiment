```python
# Creating list to append tweet data to
tweet_list = []
tweet_count = 1000
start_date = '2021-01-01'
end_date = '2021-12-08'
cryptos = ['ethereum','dogecoin']

# Using TwitterSearchScraper to scrape data and append tweets to list
dates_list = pd.Series(pd.date_range(start=start_date,end=end_date).format())

print(f'Querying {((pd.to_datetime(end_date) - pd.to_datetime(start_date)).days) * tweet_count * len(cryptos)} tweets: {tweet_count} tweets per day for {len(cryptos)} cryptos over a period of {((pd.to_datetime(end_date) - pd.to_datetime(start_date)).days+1)} days.')

for d in range(len(dates_list)-1):
    for c in cryptos: 
        for i,tweet in enumerate(sns.TwitterSearchScraper(f'{c} lang:en since:{dates_list[d]} until:{dates_list[d+1]}').get_items()):
            if i>tweet_count:
                print(f'{c} {dates_list[d]}')
                break
            tweet_list.append([tweet.date, tweet.id, tweet.user.username, c, tweet.content])

# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweet_list, columns=['Datetime', 'Tweet Id', 'Username', 'crypto', 'Text'])
```

```python
import emoji
def give_emoji_free_text(text):
    allchars = [str for str in text.encode('utf-8')]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in text.encode('utf-8').split() if not any(i in str for i in emoji_list)])
    return clean_text
```
